# La evolución de la esperanza de vida en España desde 1991 a 2020

Importamos bibliotecas necesarias:  
- pandas: para realizar los análisis de las bases de datos
- numpy: para trabajar con matrices
- matplotlib.pyplot: para trabajar con gráficas

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

Importamos las bases de datos provenientes de la INE:

```python
#DataFrame de esperanzas de vida
dataFrameEV = pd.read_csv('EsperanzaVida.csv', delimiter=';')
```

Realizamos las referencias a los datos:

```python
#DataFrame de esperanza de vida de hombres
dataFrameEVHom = dataFrameEV.head(30)[::-1]

#DataFrame de esperanza de vida de mujeres
dataFrameEVMuj = dataFrameEV.tail(30)[::-1]

#Eje x
año = dataFrameEVHom['periodo']
```

Creamos la gráfica Esperanza de Vida:

```python
#Gráfica
fig, ax = plt.subplots(figsize=(20,10))

ax.set_title('Esperanza de Vida',fontsize=30) 

ax.set_xlabel('Años', fontsize=20) 

ax.set_ylabel('Años Esperanza de Vida',fontsize=20) 

width = 0.4

plt.xticks(np.arange(1991,2020,5),fontsize=15) 

plt.yticks(np.arange(0,90,2),fontsize=10) 

ax.bar(año-0.2,dataFrameEVHom['Total'],width=width)
ax.bar(año+0.2,dataFrameEVMuj['Total'],width=width)

ax.grid(axis='y')

#Leyenda
ax.legend(['Hombres','Mujeres'], fontsize=22)
None
```
