```python
def pasarFloat(x):
    if ',' in x:
        a = x.split(',')
        ent = a[0]
        frac = a[1]
        return float(ent+'.'+frac)
    else:
        return float(x)
```

```python
df['Total 2'] = df['Total'].map(pasarFloat)
```

```python
import numpy as np
```

```python
for comunidad in df['Comunidades y Ciudades Autónomas'].unique():
    com = df['Comunidades y Ciudades Autónomas'] == comunidad
    dfcom = df[com]
    for selec in df['Nacionalidad'].unique():
        scom = dfcom['Nacionalidad'] == selec
        b = dfcom[scom]['Total 2']
        fig, ax = plt.subplots()
        ax.plot(a.values,b.values)
        ax.grid(axis='x')
        ax.grid(axis='y')
        ax.set_title(f'Evolución tasa natalidad {comunidad} ({selec})')
        plt.xticks(np.arange(2002,2021,2))
        ax.set_ylabel('‰')
        ax.set_xlabel('Año')
        None
```
