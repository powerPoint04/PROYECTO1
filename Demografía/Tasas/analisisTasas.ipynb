```python
import pandas as pd
```

```python
df = pd.read_csv('tasa paro.csv', delimiter=';')
```

```python
df
```

```python
def pasarAFloat(x):
    if x == '..':
        return 'Desconocido'
    a = x.split(',')
    return float('.'.join(a))
pasarAFloat('6,4')
```

```python
df['Porcentaje'] = df['Total'].map(pasarAFloat)
```

```python
df
```

```python
s1 = df['Comunidades y Ciudades Autónomas'] == 'Total Nacional'
s2 = df['Edad'] == 'Total'
s3 = df['Sexo'] == 'Ambos sexos'
df2 = df[s1 & s2 & s3]
df2
```

```python
porcentaje = df2['Porcentaje'].values
periodo = df2['Periodo'].values
for i in porcentaje:
    print(i)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize = (40,20))
ax.plot(periodo, porcentaje)
ax.set_title('Tasa de paro por trimestre')
labels = ax.get_xticklabels()
plt.yticks(np.arange(0,30, 2.5))
ax.grid(axis='x')
ax.grid(axis='y')
plt.setp(labels, rotation=90, horizontalalignment='right')
None
```
